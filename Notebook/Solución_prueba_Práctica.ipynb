# **4. Exploración las columnas**

Se hace una transformación a los atributos que representan un valor numérico pero se guardan de manera diferente

Vemos que a pesar de que hay variables que se considerarían numéricas, se están reconociendo como 'object' por lo que vamos a enfocarnos en estas variables.

La primera de estas variables es 'Value'.

Esta representa el valor del mercado del jugador (un valor), pero dado a la forma en la que se guarda, se reconoce como un object, por lo que se va a transformar. Esto para facilidad de la construcción de cualquier modelo y el dashboard.

```python
df['Value'].value_counts().to_frame()
```

Vemos que lo toma como un objeto dado que dentro del valor también se encuentra el signo euros y la cantidad, M para millones, K para miles.

Se procede a guardar esta variable únicamente con el valor numérico en Millones de euros.

```python
#Se aplica una función de transformación de los valores de la columna 'Value'
def label_value(row):
  value = str(row['Value'])
  length = len(value)
  num = value[1:length-1:1]
  #print(num)
  if value[0] == '€':
    if value[-1] == 'K':
      return float(num)/1000
    elif value[-1] == 'M':
      return num

df['Value'] = df.apply(lambda row: label_value(row), axis = 1)
```

```python
df['Value'].value_counts().to_frame()
```

Vemos que los valores corresponden a antes de la transformación

```python
df['Value'] = df['Value'].astype(float)
```

```python
df.rename(columns = {'Value' : 'Value_in_Million'}, inplace = True)
```

La segunda columna que vamos a transformar es 'Wage'

Esta representa el salario de un jugador (un valor), pero dado a la forma en la que se guarda, se reconoce como un object, por lo que se va a transformar. Esto para facilidad de la construcción de cualquier modelo y el dashboard.

```python
df['Wage'].value_counts().to_frame()
```

```python
#Se aplica una función de transformación de los valores de la columna 'Wage'
def label_value(row):
  wage = str(row['Wage'])
  length = len(wage)
  num = wage[1:length-1:1]
  if(wage[0]) == '€':
    if wage[-1] == 'K':
      return num
    elif wage[-1] == 'M':
      return num*1000000

df['Wage'] = df.apply(lambda row: label_value(row), axis = 1)
df['Wage'].value_counts().to_frame()
```

```python
df['Wage'] = df['Wage'].astype(float)
```

```python
df.rename(columns = {'Wage' : 'Wage_in_Thousands'}, inplace = True)
```

El cuarto atributo que se guarda como object pero representa un número es 'Height'

```python
df['Height'].value_counts().to_frame()
```

Se puede observar que no hay anomalias con los datos, por lo que se procede a transformar

```python
#Se aplica una función de transformación de los valores de la columna 'Height'
def label_value(row):
  height = row['Height'].split("'")
  #print(height)
  h = height[0] + "." + height[1]
  return float(h)
  

df['Height'] = df.apply(lambda row: label_value(row), axis = 1)
df['Height'].value_counts().to_frame()
```

De esta manera, agrupamos los valores que estaban repetidos y no se tenían en cuenta, por ejemplo, el valor "5'1" y "5'10" que son el mismo, al cambiar su representación a un float, se toman ahora si como el mismo valor, esto no pasaba antes de hacer la transformación.

El quinto atributo que representa un número pero el DF lo guarda como un 'objec' es 'Weight'.

```python
df['Weight'].value_counts().to_frame()
```

Se puede observar que esto es debido a que se guardan las unidades dentro. Gracias a la anterior tabla se puede ver que la unidad siempre es la misma, es decir, libras. Por lo que se procede a transformar este valor.

```python
#Se aplica una función de transformación de los valores de la columna 'Weight'
def label_value(row):
  weight = str(row['Weight'])
  w = weight[0:-3:1]
  #print(w)
  return int(w)
  
df['Weight'] = df.apply(lambda row: label_value(row), axis = 1)
df['Weight'].value_counts().to_frame()
```
